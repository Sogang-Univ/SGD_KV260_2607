# Custom AXI-lite GPI Control

## Import Overlay

```python
from pynq import Overlay
```

## Overlay bitstream

### load `.bit` and`.hwh` files and then overlay

```python
bitstream_name = 'prj_01_05.bit'

try:
    custom_ol = Overlay(bitstream_name)
    print(f"{bitstream_name} 로드 완료.")
except FileNotFoundError:
    print(f"오류: {bitstream_name} 파일을 찾을 수 없습니다.")
    exit()
```

## Get the module instance

### check the module name

```python
# 로드된 모든 IP 이름 확인
print("로드된 IP 리스트:")
print(custom_ol.ip_dict.keys())
```

### get the module instance

```python
# 1. 사용자 정의 IP 객체 가져오기
gpio_axi = getattr(custom_ol, 'AXI4_lite_s_0')
gpio_axi
```

## Enjoy with the module

### set registers

```python
AXI_DATA_BYTE = 4

BaseAddress = 0 #0xA0000000
Reg_key = BaseAddress + AXI_DATA_BYTE * 0
Reg_seg = BaseAddress + AXI_DATA_BYTE * 1
```

### register write

```python
gpio_axi.write(Reg_seg, 0xfe)
```

### register read

```python
gpio_axi.read(Reg_seg)
```
